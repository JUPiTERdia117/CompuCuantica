## Creando circuito

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Operator
from qiskit.circuit.library import UnitaryGate
from qiskit.visualization import plot_histogram
from qiskit.providers.basic_provider import BasicSimulator

circuit = QuantumCircuit(3) # circuito con 3 qubits

circuit.h(2)    # Hadamard en q2
circuit.cx(1,2)   # CNOT Con q1 como control y q2 como obj
circuit.tdg(2)   # Aplica T inversa a q2

circuit.cx(0,2)   # CNOT Con q0 como control y q2 como obj

circuit.t(2)   # Aplica T a q2
circuit.cx(1,2)   # CNOT Con q1 como control y q2 como obj

circuit.tdg(2)   # Aplica T inversa a q2

circuit.cx(0,2)   # CNOT Con q0 como control y q2 como obj

circuit.barrier()

circuit.t(1)   # Aplica T  a q1
circuit.t(2)   # Aplica T  a q2

circuit.cx(0,1)   # CNOT Con q0 como control y q1 como obj

circuit.h(2)    # Hadamard en q2

circuit.t(0)   # Aplica T  a q0
circuit.tdg(1)   # Aplica T inversa a q1

circuit.cx(0,1)   # CNOT Con q0 como control y q1 como obj



#Desplegamos el circuito.
display(circuit.draw('mpl'))

#Vemos al operador como matriz.
op = Operator(circuit)
display(array_to_latex(op))


#Guardamos al estado final del circuito.
#psi = Statevector(circuit)
#Desplegamos al estado represnetado en la esfera de Boch.
#display(plot_bloch_multivector(psi))
```

Para más fexibilidad guardo la matriz en un circuito personalizado para aplicarlo varias veces

```python
matrix = op

mi_gate = UnitaryGate(matrix, 'mi_gate') #Compuerta personalizada

```

## Aplicando circuito

Con la compuerta personalizada guardada se puede aplicar a los distintos circuitos


### Circuito $\lvert 000\rangle$

```python
circ000 = QuantumCircuit(3, 3)      # Circuito 000

estado_000 = Statevector(circ000)   # Estado 000
print("➤ Estado inicial en forma de ket:")
display(estado_000.draw('latex'))


circ000.append(mi_gate, [0,1,2])     # le agregamos al circuito la compuerta personalizada

res_000 = Statevector(circ000)   # Estado resultante después de aplicar compuerta personalizada

circ000.measure([0,1,2], [0,1,2])  # medir los 3 qubits en los 3 bits clásicos

print("➤ Estado resultado en forma de ket:")
display(res_000.draw('latex'))

display(circ000.draw('mpl'))
```

## Tabla de verdad 

Con los resultados obtenidos podemos concluir que la tabla de verdad del circuito es la siguiente:
|q2|q1|q0|$\rightarrow$|q2|q1|q0|
|:----:|:----:|:----:|:----:|:----:|:----:|:----:|
|0|0|0|$\rightarrow$|0|0|0|
|0|0|1|$\rightarrow$|0|0|1|
|0|1|0|$\rightarrow$|0|1|0|
|0|1|1|$\rightarrow$|1|1|1|
|1|0|0|$\rightarrow$|1|0|0|
|1|0|1|$\rightarrow$|1|0|1|
|1|1|0|$\rightarrow$|1|1|0|
|1|1|1|$\rightarrow$|0|1|1|



## ¿Qué hace?

El circuito solo parece funcionar cuando q1 y q0 son iguales a 1, y si esto sucede aplican una compuerta X al qbit q2.
Lo que significa que q0 es control de q1, y q1 es control de q2. Esto es la compuerta toffoli aplicada a q0 y q1 como controles
y a q2 como objetivo.
